# Growth Visualization

The following file is meant to display the trends that exist within each of the available datasets. In this file, the we will be examining these different trends through the perspective of the five different regions of the United States, defined by official U.S as the following:



**Northeast Region (States: Connecticut, Maine, Massachusetts, New Hampshire, Rhode Island, Vermont, New Jersey, New York, Pennsylvania)** 

**Midwest Region (States: Indiana, Illinois, Michigan, Ohio, Wisconsin, Iowa, Nebraska, Kansas, North Dakota, Minnesota, South Dakota,Missouri)**

**South Region (States: Delaware, Florida, Georgia, Maryland, North Carolina, South Carolina, Virginia, West Virginia, Alabama, Kentucky,Mississippi, Tennessee, Arkansas, Louisiana, Oklahoma, Texas)**

**West Region (States: Arizona, Colorado, Idaho, New Mexico, Montana, Utah, Nevada, Wyoming, California, Oregon, Washington)**

**Alaska Region: (States: Alaska)**
**Hawaii Region: (States: Hawaii)**



For the purposes of being considered a territory, we will not be including the District of Columbia in our calculation of the South. In addition to this, we will be defining the states of Alaska and Hawaii as their own regions, due to their lack of presence on the continental United States. It would not make sense to use the data for Alaska or Hawaii to calculate the Regional Growth Index for a state like Wyoming, since they are so far away. 

```python
# Importing Packages Use for Visualization

import matplotlib.pyplot as plt
import scipy.stats as scs
import statsmodels.stats.api as sms
import statsmodels.api as sm
import pandas as pd
import numpy as np
import matplotlib.ticker as ticker
```

```python
US_Census_df = pd.read_csv('United_States_Census_Data.csv')
US_Fertility_df = pd.read_csv('United_States_Fertility_Rates.csv')
```

# Census Visualizations

```python
formatter = ticker.FuncFormatter(lambda x, pos: '{:,}'.format(int(x)))
pop_years = ['1910', '1920', '1930', '1940', '1950', '1960', '1970', '1980', '1990', '2000', '2010', '2020']

def region_condenser_pop(idx_lst):
    regional_data = [0 for i in range(12)]

    for i in idx_lst:
        data = US_Census_df.iloc[i].iloc[2:]  
    
        data_as_integers = data.str.replace(',', '').astype(int)

        storage = data_as_integers.tolist()
        storage.reverse()

        regional_data = [sum(i) for i in zip(regional_data, storage)]
    
    return regional_data
```

### Northeast Regional Population Data Visualization

```python
northeast_indices = [6, 18, 20, 28, 38, 44, 39, 31, 37]
```

```python
plt.figure(figsize=(10, 10))
plt.plot(pop_years, region_condenser_pop(northeast_indices))
plt.xlabel('Years of Census')
plt.gca().xaxis.set_major_locator(ticker.MultipleLocator(base=2))
plt.ylabel('Number of Regional Residents')
plt.title('Northeast Population over Years')
plt.grid(True)
plt.gca().yaxis.set_major_formatter(formatter)


directory_path = '/Users/emmanuely/Desktop/CodeFolder/Region_Growth_Projections/Figure_visualizations/Northeast_region_visuals/'

plt.savefig(directory_path + 'Northeast_region_population.jpeg')
plt.show()  
```
